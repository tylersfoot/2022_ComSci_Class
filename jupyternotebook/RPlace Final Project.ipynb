```python
import datetime
import time
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update(plt.rcParamsDefault)
%matplotlib inline
df0 = pd.read_csv('./data/place/place0.csv', sep=',')
df70 = pd.read_csv('./data/place/place70.csv', sep=',')
# df71 = pd.read_csv('./data/place/place71.csv', sep=',')
# df72 = pd.read_csv('./data/place/place72.csv', sep=',')
# df73 = pd.read_csv('./data/place/place73.csv', sep=',')
# df74 = pd.read_csv('./data/place/place74.csv', sep=',')
# df75 = pd.read_csv('./data/place/place75.csv', sep=',')
# df76 = pd.read_csv('./data/place/place76.csv', sep=',')
# df77 = pd.read_csv('./data/place/place77.csv', sep=',')
# df78 = pd.read_csv('./data/place/place78.csv', sep=',')
# imports libraries and loads in dataset
# df = pd.concat([df70,df71,df72,df73,df74,df75,df76,df77,df78], ignore_index=True, axis=0, sort=False)
df = df0
# df = df[:5000]
df # shows dataset

# 83 hours
# new colors added every day
# 2000 x 1000 around 16:25 GMT on April 2
# 2000 x 2000 around 19:00 GMT on April 3
```

```python
# fixes coordinates
def split_coords(row):
    # converts a coordinate (34, 21) into two integers, 34 and 21
    x = row[:row.find(',')]
    y = row[row.find(',')+1:]
    if y != y[:row.find(',')+2]:
        y = y[:row.find(',')]
    return {'x' : int(x), 'y' : int(y)}

dftemp = df["coordinate"].apply(lambda x: split_coords(x)) # temporary dataframe to store x and y values
dftemp = pd.DataFrame(list(dftemp))
df = pd.concat([df, dftemp], axis=1, ignore_index=True) # merges the temporary dataframe with the main dataframe
df = df.set_axis(['time', 'user', 'color', 'coordinate', 'x', 'y'], axis=1, inplace=False) # renames columns
df.head(20)
```

```python
# converts time to unix
format = "%Y-%m-%d %H:%M:%S"
def time_conversion(row):
    row = row[:row.find(':')+6]
    date_time = datetime.datetime.strptime(row, format)
    return time.mktime(date_time.timetuple())

dftemp = df["time"].apply(lambda x: time_conversion(x)) # temporary dataframe to store unix values
dftemp = pd.DataFrame(list(dftemp))
df = pd.concat([df, dftemp], axis=1, ignore_index=True) # merges the temporary dataframe with the main dataframe
df = df.drop([0, 3], axis=1)
df = df.set_axis(['user', 'color', 'x', 'y', 'time'], axis=1, inplace=False) # renames columns
df.head(20)
```

```python
dfColors = df[['color','user']].groupby('color').count() # dataframe with each color and its count
dfColors.reset_index(inplace=True)
dfColors = dfColors.set_axis(['color', 'count'], axis=1, inplace=False) # renames columns
dfColors = dfColors.sort_values(by='count', ascending=False) # sorts by count highest to lowest
dfColors.head(5)
```

```python
countList = dfColors['count'].tolist()
clrs = ['grey' if (x < max(countList)) else 'red' for x in countList]
for i in range(len(clrs)):
    clrs[i] = dfColors['color'].iloc[i] # sets each bar color to its respective hex value
plt.style.use('seaborn-dark')
sns.set(rc={'axes.facecolor':'lightgray', 'figure.facecolor':'lightgray'}) # sets background colors
fig, ax = plt.subplots(figsize=(15, 10))
colorsBar = sns.barplot(data=dfColors, x='count', y='color', ci=0.9, ax=ax, palette=clrs, edgecolor='black', linewidth=0)
plt.title('Number of Pixels Placed per Color', fontsize=20)
plt.xlabel('Count', fontsize=20)
plt.ylabel('Color', fontsize=20)
plt.show()
```

```python
dfUsers = df[['user','color']].groupby('user').count() # dataframe with each user and their # of pixels placed
dfUsers.reset_index(inplace=True)
dfUsers = dfUsers.set_axis(['user', 'count'], axis=1, inplace=False) # renames columns
dfUsers = dfUsers.sort_values(by='count', ascending=False) # sorts by count highest to lowest
dfUsers.head(5)
```

```python
plt.style.use('seaborn-dark')
sns.set(rc={'axes.facecolor':'lightgray', 'figure.facecolor':'lightgray'}) # sets background colors
fig, ax = plt.subplots(figsize=(15, 10))
usersBar = sns.lineplot(data=dfUsers, x='user', y='count', ci=0.9, ax=ax)
plt.title('Number of Pixels Placed per User', fontsize=20)
plt.xlabel('User', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.show()
```
